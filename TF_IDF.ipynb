```python
import urllib.request
import bs4 as BeautifulSoup
import nltk
import re
nltk.download('punkt')
from nltk.corpus import stopwords
from string import punctuation
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize 
```

```python
data = urllib.request.urlopen('https://en.wikipedia.org/wiki/Artificial_intelligence')
url='https://en.wikipedia.org/wiki/Artificial_intelligence'
request = urllib.request.urlopen(url).read().decode('utf8','ignore')
```

```python
note_parsed = BeautifulSoup.BeautifulSoup(request,'html.parser')
```

```python

para = note_parsed.find_all('p')
```

```python
note_content = ''
for p in para:  
    note_content += p.text
```

```python
note_content
```

```python
for line in note_content:
         x=line.strip();
         print(x)
```

```python
note_content = re.sub(r'[[0-9]*]', ' ', note_content)
```

```python
formatted_note_content = re.sub('[^a-zA-Z]', ' ', note_content)
```

```python
print (note_content)
```

```python
sent = sent_tokenize(note_content)
sent
```

```python
tokens = word_tokenize(note_content)
```

```python
tokens
```

```python
nltk.download("stopwords")
st_w=stopwords.words('english')
```

```python
freq = {}
for word in tokens:    
    if word.lower() not in st_w:
        if word.lower() not in punctuation:
            if word not in freq.keys():
                   freq[word] = 1
            else:
                   freq[word] += 1
```

```python
freq
```

```python
max_freq = max(freq.values())
print(max_freq)
```

```python
for word in freq.keys():
      freq[word] = freq[word]/max_freq
```

```python
print(freq)
```

```python
sentence_scores = {}
for sente in sent:
    for word in nltk.word_tokenize(sente.lower()):
        if word in freq.keys():
            if len(sente.split(' ')) < 30:
                if sente not in sentence_scores.keys():
                    sentence_scores[sente] = freq[word]
                else:
                    sentence_scores[sente] += freq[word]
```

```python
print(sentence_scores)
```

```python
import heapq
summary_sentences = heapq.nlargest(7, sentence_scores, key=sentence_scores.get)
summary_sentences
```
